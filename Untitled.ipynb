```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stent = pd.read_csv('data/stent30.csv')
stent = stent.rename(columns = {'outcome': 'outcome30'})
stent['outcome365'] = pd.read_csv('data/stent365.csv', usecols = ['outcome'])
stent.head()
```

```python
# Number of columns and rows
print('Data Matrix shape -- ' + 'Rows: ' + str(stent.shape[0]) + ' Columns: ' +  str(stent.shape[1]))
```

```python
# Number of no events and stroke for Treatment and control group
print('Treatment: \n',stent[stent['group'] == 'treatment']['outcome30'].value_counts())
print('\n')
print('Control: \n',stent[stent['group'] == 'control']['outcome30'].value_counts())
```
